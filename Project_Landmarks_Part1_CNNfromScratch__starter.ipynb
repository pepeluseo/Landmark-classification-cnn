Given the large number of landmarks across the world and the immense volume of images that are uploaded to photo sharing services, using human judgment to classify these landmarks would not be feasible.

For preprocessing, all images are first resized to 256 pixels to ensure a consistent input scale. For the training set, I then apply a random crop to 224x224 pixels, followed by random horizontal flipping and random rotation. These augmentation steps expose the model to slightly different versions of the same landmark images, which helps reduce overfitting and improves generalization.

For the validation and test sets, I use a center crop instead of random transformations. This makes evaluation deterministic and consistent, because the validation and test images are not randomly modified each time they are loaded. Finally, all images are converted to tensors and normalized using the dataset mean and standard deviation. Normalization helps stabilize the training process and allows the CNN to learn more effectively.

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__ 

I designed the CNN using a sequence of convolutional blocks followed by a fully connected classifier. Each convolutional block contains a convolutional layer, batch normalization, ReLU activation, and max pooling. The early convolutional layers learn simple visual patterns such as edges and textures, while deeper layers can learn more complex landmark-specific structures.

Max pooling reduces the spatial size of the feature maps, making the network more computationally efficient and helping it focus on the most relevant features. Batch normalization helps stabilize training, while ReLU introduces non-linearity. I also use dropout in the classifier to reduce overfitting, which is important because the model is trained from scratch.

The final linear layer uses the `num_classes` parameter, so the model output automatically matches the number of landmark categories. I do not apply Softmax in the forward method because `CrossEntropyLoss` expects raw logits.

Finally, let's have a look at the confusion matrix of the model we are going to use in production:

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

class_names = data_loaders["test"].dataset.classes

pred_np = np.array([int(p) for p in pred])
truth_np = np.array([int(t) for t in truth])

present_classes = np.unique(np.concatenate([truth_np, pred_np]))

cm = confusion_matrix(
    truth_np,
    pred_np,
    labels=present_classes
)

display_labels = [class_names[i] for i in present_classes]

fig, ax = plt.subplots(
    figsize=(max(8, len(present_classes) * 1.5), max(6, len(present_classes) * 1.2))
)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=display_labels
)

disp.plot(
    ax=ax,
    cmap="Blues",
    colorbar=False,
    values_format="d",
    xticks_rotation=45
)

plt.title("Confusion Matrix - Classes Present in Test Predictions")
plt.tight_layout()
plt.show()
````
